Design a box with the following functionalities and ensure they all work independently and in any combination.
The right button click on a mouse will generate random movable points with different colours going in any random direction diagonally within a boundary region. For instance, if a point is generated at (0,0), it can go to (-1, 1), (-1, -1), (1,1), or (1, -1), and so on. The points should be spawned where the right button click will be given in the box and the colour and direction of movement should be random. The points will continue to move in the same direction and will bounce back from the wall of the boundary.
Pressing the “up arrow” key on the keyboard will increase the speed of all the points generated so far and pressing the “down arrow” key on the keyboard will decrease the speed.
The left button click on a mouse will make the points blink i.e. if a point is in red, it will go background color(here it’s black) and return to red, and this transition should take place within a second while the transition cycle goes on.
Pressing the “Spacebar” on the keyboard should freeze all the points and none of the above functionalities will work when frozen. The same “Spacebar” should unfreeze them.

```python
from OpenGL.GL import *
from OpenGL.GLUT import *
from OpenGL.GLU import *
import math
width = 500
height = 500
points = []
speed = 2
frozen = False
blink = False
def draw_points():
    global blink
    for i in points:
        x = i[0]
        y = i[1]
        direction_x = i[2]
        direction_y = i[3]
        color = i[4]
        if(blink == True):
            if(int(glutGet(GLUT_ELAPSED_TIME) / 500) % 2 == 0):
                glColor3f(0.0, 0.0, 0.0)
            else:
                glColor3f(*color)
        else:
            glColor3f(*color)
        glBegin(GL_POINTS)
        glVertex2f(x, y)
        glEnd()
def update_points():
    global points
    global speed
    for i in points:
        i[0] += i[2] * speed
        i[1] += i[3] * speed
        if(i[0] <= 0 or i[0] >= width):
            i[2] *= -1
        if(i[1] <= 0 or i[1] >= height):
            i[3] *= -1
def mouse(button, state, x, y):
    global points
    if(state == GLUT_DOWN):
        opengl_x = x
        opengl_y = height - y
        if(button == GLUT_RIGHT_BUTTON):
            direction_x = (x % 2) * 2 - 1
            direction_y = (y % 2) * 2 - 1
            color = [(x % 255) / 255.0,(y % 255) / 255.0,((x + y) % 255) / 255.0,]
            points.append([opengl_x, opengl_y, direction_x, direction_y, color])
        elif(button == GLUT_LEFT_BUTTON):
            global blink
            blink = not blink
def keyboard(key, x, y):
    global speed
    global frozen
    if(key == b' '):
        frozen = not frozen
    elif(key == b'\x1b'):
        glutLeaveMainLoop()
def keyboard_updown(key, x, y):
    global speed
    if(key == GLUT_KEY_UP):
        speed += 1
    elif(key == GLUT_KEY_DOWN):
        speed = max(1, speed - 1)
def idle():
    if(frozen == False):
        update_points()
    glutPostRedisplay()
def display():
    glClear(GL_COLOR_BUFFER_BIT | GL_DEPTH_BUFFER_BIT)
    draw_points()
    glutSwapBuffers()
def init():
    glMatrixMode(GL_PROJECTION)
    glLoadIdentity()
    gluOrtho2D(0, width, 0, height)
    glPointSize(10)

glutInit()
glutInitDisplayMode(GLUT_DOUBLE | GLUT_RGB | GLUT_DEPTH)
glutInitWindowSize(width, height)
glutCreateWindow(b"Task 2: Building the Amazing Box")
init()
glutDisplayFunc(display)
glutMouseFunc(mouse)
glutKeyboardFunc(keyboard)
glutSpecialFunc(keyboard_updown)
glutIdleFunc(idle)
glutMainLoop()
```
